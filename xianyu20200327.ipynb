```python
hs300 = raw_data[['300ETF','signal']].copy()
```

```python
hs300['Close'] = raw_data['300ETF']
hs300['Open'] = raw_data['300ETF']
hs300['High'] = raw_data['300ETF']
hs300['Low'] = raw_data['300ETF']
```

```python
def get_data(values):
    return pd.Series(values)
```

```python
class Strategy1(Strategy):
    
    def init(self):
        self.close = self.I(get_data,self.data.Close)
        self.signal = self.I(get_data,self.data.signal)
        #self.number = self.I(get_data,self.data.number)
    def next(self):
        #print(self.number[-1],self.signal[-1])
        if self.signal==1:
            self.buy()
            self.pos=1
            
        elif self.signal==0:
            self.sell(self.close[-2])
            self.pos=0
```

```python
bt = Backtest(raw_data, Strategy1, cash=1000000, commission=0)

output = bt.run()
bt.plot()
```
